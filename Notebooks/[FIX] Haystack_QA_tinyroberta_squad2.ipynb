```python
from haystack.nodes import PDFToTextConverter,PreProcessor
from haystack.preprocessor import PreProcessor
#from haystack.document_stores.faiss import FAISSDocumentStore
from haystack.retriever import DensePassageRetriever
from haystack.reader import FARMReader
from haystack.pipeline import ExtractiveQAPipeline
```

```python
doc = 'test.pdf'
```

```python
exists(doc)
```

```python
pdf_converter = PDFToTextConverter(remove_numeric_tables=True, valid_languages=["en"])
converted = pdf_converter.convert(file_path = 'test.pdf', meta = { "company": "Company_1", "processed": False })
```

```python
preprocessor = PreProcessor(split_by="word", split_length=200, split_overlap=10)
preprocessed = preprocessor.process(converted)
```

```python
document_store = FAISSDocumentStore(faiss_index_factory_str="Flat", return_embedding=True)
document_store.delete_all_documents()
document_store.write_documents(preprocessed)
```

```python
retriever = DensePassageRetriever(document_store=document_store)
reader = FARMReader(model_name_or_path='deepset/tinyroberta-squad2', use_gpu=True)
document_store.update_embeddings(retriever)
```

```python
pipeline = ExtractiveQAPipeline(reader, retriever)
```

```python
%%time
questions = 'what is the basic approach to conversion optimisation?'

prediction = pipeline.run(query=questions, params={"Retriever": {"top_k": 5}, "Reader": {"top_k": 10}})
```

```python
print('Top 5 Answers: ')

for i in range(5):
  print('Answer Number', i+1)
  print('Q:', prediction['query'])
  print('A:', prediction['answers'][i].answer)
  print('Context:', prediction['answers'][i].context)
  print('score: ',prediction['answers'][i].score)
  print('\n')
```
